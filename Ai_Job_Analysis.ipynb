## Visualization and Analysis

```python
if 'Job Title' in df.columns:
    High_number_jobs = df['Job Title'].value_counts().head(10)
    print(High_number_jobs)
```

```python
mat.figure(figsize=(10,6))
High_number_jobs.plot(kind='bar' , x = High_number_jobs , y = High_number_jobs.index , width=0.2 , edgecolor='black')
mat.title('Top 10 Highest Jobs ')
mat.xlabel('Job Title')
mat.ylabel('Number of Jobs')
mat.grid(axis='y', linestyle='--', alpha=0.5)
mat.show()
```

```python
if 'Job Status' in df.columns:
    Status = df['Job Status'].value_counts()
    print(Status)
```

```python
Status.plot(kind='bar' , x = Status , y = Status.index , width=0.4 , color=['green','red'] )
mat.title('Total Number of Job Status')
mat.xlabel('Job Status')
mat.ylabel('Value Counts')
mat.grid(axis='y', linestyle='--', alpha=0.5)
mat.show()
```

```python
top_jobs = df.groupby('Job Title')['Median Salary (USD)'].mean().nlargest(10)
top_jobs
```

```python
mat.figure(figsize=(10,6))
top_jobs.plot( kind = 'bar', color='teal',edgecolor='black', width=0.5 )
mat.title('Top 10 Jobs with Average Salaries')
mat.xlabel('Job Title')
mat.ylabel('Median Salary(USD)')
mat.grid(axis='y', linestyle='--', alpha=0.5)
mat.show()

```

```python
top_industries = df.groupby('Industry')['Median Salary (USD)'].mean().sort_values(ascending=False).nlargest(10)
top_industries
```

```python
mat.figure(figsize=(10,6))
top_industries.plot( kind = 'bar', color='teal',edgecolor='black', width=0.5 )
mat.title('Top 10 Industries by Average Salaries')
mat.xlabel('Industry')
mat.ylabel('Median Salary(USD)')
mat.grid(axis='y', linestyle='--', alpha=0.5)
mat.show()

```

```python
sns.countplot(data=df, x='AI Impact Level',edgecolor='black', order=df['AI Impact Level'].value_counts().index)
mat.title("Distribution of AI Impact Levels")
mat.grid(axis='y', linestyle='--', alpha=0.5)
mat.show()

```

```python
df['Growth (%)'] = ((df['Projected Openings (2030)'] - df['Job Openings (2024)']) / df['Job Openings (2024)']) * 100
df
```

```python
mat.figure(figsize=(10,8))
sns.barplot(data=df.sort_values('Growth (%)', ascending=False).head(10),x='Job Title', y='Growth (%)', hue='Industry',edgecolor='black')
mat.title("Top 10 Jobs by Projected Growth (%)")
mat.xticks(rotation=90)
mat.grid(axis='y', alpha=0.5)
mat.legend(bbox_to_anchor=(1,0.5),shadow=True,frameon=True,facecolor='lightyellow',edgecolor='black')
mat.show()
```
